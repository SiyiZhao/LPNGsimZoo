```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
fastpm_input = np.loadtxt('../fastpm-ic/Planck2015Table4LastColumn_matterpower.dat')
```

```python
k1 = fastpm_input[:, 0]
pk1 = fastpm_input[:, 1]
```

```python
ic2lpt_input = np.loadtxt('../ic-2LPT/log10pkl_z0.dat')
k2 = 10**ic2lpt_input[:, 0]
pk2 = 10**ic2lpt_input[:, 1]
```

```python
ic2lpt_tk = np.loadtxt('../ic-2LPT/Tk_z0.dat')
k3 = ic2lpt_tk[:, 0]
tk = ic2lpt_tk[:, 1]
pk3 = tk*tk*k3
```

```python
plt.plot(k1, pk1)
plt.plot(k2, pk2, '--')
plt.plot(k3, pk3, ':')
plt.legend(['FastPM', 'IC2LPT', 'IC2LPT Tk'])
plt.xlabel(r'$k$ [h/Mpc]')
plt.ylabel(r'$P(k)$ [Mpc/h]$^3$')
plt.xscale('log')
plt.yscale('log')
```
